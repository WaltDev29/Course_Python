# DataFrame 결합

```python
import pandas as pd

df1 = pd.DataFrame({"name" : ["Jessica", "Liam", "Sophia", "Alex"],
				   "age": [10,30,29,14]})

df2 = pd.DataFrame({"name" : ["Jessica", "Liam", "Sophia", "Alex"],	
					"department": ["AI", "Law", "AI", "Law"],		   
				   "score": [100,70,80,95],})

df3 = pd.DataFrame({"department": ["AI", "Law"],
					"head" : ["Steve", "Emma"]})
```

## 1. merge()

df + df 가로로 병합

### 1-1. 공통열이 하나일 경우

```python
df = pd.merge(df1, df2)
display(df)
```

### 1-2. 공통 열이 여러 개일 경우

on 파라미터에 기준 열을 지정  
**공통 열이 하나여도 명시적으로 지정해주는 것이 바람직함**

```python
df2["age"] = [10,20,30,40]
df = pd.merge(df1,df2, on="name")
display(df)
```

이런 것도 가능

```python
display(pd.merge(df, df3, on="department"))
```

## 2. join()

merge()보다 간단함.  
인덱스 기준으로 병합.

```python
df1 = pd.DataFrame({"age":[10,20,30,40]})
df2 = pd.DataFrame({"gender": ['F','M','M','F']})
df = df1.join(df2)

display(df)
```

## 3. concat()

기존 DataFrame에 df를 추가  
**ignore_index** : index를 다시 매길 건지 (기본값 : False)

```python
df1 = pd.DataFrame({"name" : ["Jessica", "Liam", "Sophia", "Alex"],
				   "age": [10,30,29,14]})

df2 = pd.DataFrame({"name": ["James", "Eric"],
					"age": [20,30]})

display(pd.concat([df1,df2]))
display(pd.concat([df1, df2], ignore_index=True))
```
