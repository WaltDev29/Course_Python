# pandas 데이터 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({"name" : ["Jessica", "Liam", "Sophia", "Alex"],
				   "age": [10,30,29,14],
				   "assign1": [100,70,80,95],
				   "assign2": [60,30,50,70],
				   "test": [10,20,30,40]})
```

## 1. 막대 그래프

```python
plt.figure(figsize=(4,3))
plt.bar(df["name"], df["assign1"])
plt.ylabel("Score")
plt.show()
```

## 2. 누적 막대 그래프

**bottom** : 해당 bar 그래프를 밑으로 설정

```python
plt.figure(figsize=(4,3))
plt.bar(df["name"], df["assign1"], label="assign1")
plt.bar(df["name"], df["assign2"], bottom=df["assign1"], label="assign2")
plt.bar(df["name"], df["test"], bottom=df["assign1"] + df["assign2"], label="test")
plt.legend()
plt.ylabel("Score")
plt.show()
```
