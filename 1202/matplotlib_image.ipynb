# matplotlib로 이미지 다루기

```python
import matplotlib.pyplot as plt
```

## 1. 이미지 다루기

```python
cat = plt.imread('imgs/cat1.jpg')
dog = plt.imread('imgs/dog1.jpg')
imgs = [cat, dog]
labels = ['cat', 'dog']

plt.figure(figsize=(5, 3))
for i in range(len(imgs)):
    plt.subplot(1, 2, i+1)
    plt.imshow(imgs[i])
    plt.xticks([]) # x축 눈금 제거
    plt.yticks([]) # y축 눈금 제거
    plt.xlabel(labels[i]) # 이미지 아래 제목

plt.show()
```

## 2. os.listdir() 이용

```python
import os

imgs = os.listdir('./imgs') # 'imgs' 폴더 내 파일 목록 가져오기

plt.figure(figsize=(5, 5))
for i in range(len(imgs)):
    plt.subplot(3, 3, i+1) # 3행 3열
    img = plt.imread(os.path.join('./imgs', imgs[i])) # 경로 결합
    plt.imshow(img)
    plt.axis('off') # axis off를 사용하면 title 설정 불가

plt.show()
```
